```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
x=np.random.rand(100,2)
```

```python
x
```

```python
plt.scatter(x[:,0],x[:,1])
```

```python
clf=KMeans(n_clusters=70)
```

```python
clf.fit(x)
```

```python
clf.labels_
```

```python
plt.scatter(x[:,0],x[:,1],c=clf.labels_)
```
