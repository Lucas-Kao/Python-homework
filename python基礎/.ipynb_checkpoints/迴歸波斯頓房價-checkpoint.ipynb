```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.datasets import load_boston
```

```python
b=load_boston()
```

```python
b.feature_names
```

```python
x=b.data
y=b.target
```

```python
len(x)
```

```python
len(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.7,test_size=0.3,random_state=20)
```

```python
r=LinearRegression()
```

```python
r.fit(x_train,y_train)
```

```python
y_pre=r.predict(x_test)
```

```python
plt.scatter(y_pre,y_test)
plt.plot([0,50],[0,50],"pink")
plt.xlabel("predict price")
plt.ylabel('real price')
```

```python
plt.ylabel("actual price")
```

```python
plt.figure(figsize=(12,12))
for i,feature in enumerate(b.feature_names):
    plt.subplot(5,3,i+1)
    plt.scatter(x[:,i],y,s=0.1)
    plt.ylabel("price")
    plt.xlabel(feature)
    plt.tight_layout()
```
